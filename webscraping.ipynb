```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url='https://books.toscrape.com/catalogue/category/books_1/page-1.html'
```

```python
response=requests.get(url)
```

```python
if response.status_code==200:
  print("sccessfully accessed")
else:
  print('failed')
```

```python
soup= BeautifulSoup(response.text,'html.parser')
```

```python
books=[]
```

```python
for h3 in soup.find_all('h3'):
  title =h3.find('a').get('title')
  books.append(title)
print(books)
```

```python
url='https://books.toscrape.com/catalogue/category/books_1/page-{}.html'
```

```python
all_books=[]
for page_num in range(1,51):
  url= url.format(page_num)
  response =requests.get(url)
  if response.status_code ==200:
    print(f"page successfully accessed{page_num}")
  else:
    print(f"page access denied{page_num}")
    continue
  soup=BeautifulSoup(response.text,'html.parser')
  books = soup.find_all("article", class_="product_pod")
  for book in books:
    title=  book.h3.a['title']
    price= book.find('p',class_='price_color').text
    availability= book.find('p',class_='instock availability').text.strip()
    rating = book.p['class'][1]
    all_books.append([title,price,availability,rating])
print(all_books)
```

```python
col=['Title','Price','Availability','Rating']
df=pd.DataFrame(all_books,columns=col)
df
```

```python
df.to_csv('book.csv',index=False,encoding='utf-8') # download data
```

```python
df.info()
```

```python
df['Price']=df['Price'].str.replace(r'[^\d.]','',regex=True)
```

```python
df.info()
```

```python
df['Price']=df['Price'].astype('float')
```

```python
df.info()
```

```python
mapping={'One':1,
         'Two':2,
         'Three':3,
         'Four':4,
         'Five':5}
```

```python
df['Rating']=df['Rating'].map(mapping)
```

```python
df
```

```python
df.info()
```

```python
df['Availability']=df['Availability'].apply(lambda x:'In stock' in x)
```
